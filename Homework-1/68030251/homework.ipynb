```
///ข้อที่1
// นับเลข 1-100
Console.WriteLine("นับเลข 1-100:");
for (int i = 1; i <= 100; i++)
{
    Console.Write(i + " ");
}
Console.WriteLine("\n");
// แสดงเฉพาะเลขคี่
Console.WriteLine("เลขคี่ 1-100:");
for (int i = 1; i <= 100; i++)
{
    if (i % 2 != 0)
    {
        Console.Write(i + " ");
    }
}
Console.WriteLine("\n");
// แสดงเฉพาะเลขคู่
Console.WriteLine("เลขคู่ 1-100:");
for (int i = 1; i <= 100; i++)
{
    if (i % 2 == 0)
    {
        Console.Write(i + " ");
    }
}
Console.WriteLine("\n");
// แสดงเฉพาะเลขจำนวนเฉพาะ
Console.WriteLine("เลขจำนวนเฉพาะ 1-100:");
for (int i = 2; i <= 100; i++)
{
    bool isPrime = true;
    for (int j = 2; j <= Math.Sqrt(i); j++)
    {
        if (i % j == 0)
        {
            isPrime = false;
            break;
        }
    }
    if (isPrime)
    {
        Console.Write(i + " ");
    }
}
Console.WriteLine();

```

```
//ข้อ2
Console.Write("กรุณากรอกอายุ: ");
int age = int.Parse(Console.ReadLine());

if (age >= 1 && age <= 12)
{
    Console.WriteLine("เด็ก");
}
else if (age >= 13 && age <= 19)
{
    Console.WriteLine("วัยรุ่น");
}
else if (age >= 20 && age <= 50)
{
    Console.WriteLine("วัยผู้ใหญ่");
}
else if (age >= 51)
{
    Console.WriteLine("วัยชรา");
}
else
{
    Console.WriteLine("กรุณากรอกอายุที่ถูกต้อง");
}
```
