```
///นับเลข 1 – 100
for (int i = 1; i <= 100; i++)
{
    Console.WriteLine(i);
}
```

```
///เลขคี่
for (int i = 1; i <= 100; i++)
{
    if (i % 2 != 0)
    {
        Console.WriteLine(i);
    }
}
```

```
///เลขคู่
for (int i = 1; i <= 100; i++)
{
    if (i % 2 == 0)
    {
        Console.WriteLine(i);
    }
}
```

```
///แสดงเฉพาะเลขจำนวนเฉพาะ
for (int i = 2; i <= 100; i++)
{
    bool isPrime = true;
    for (int j = 2; j <= Math.Sqrt(i); j++)
    {
        if (i % j == 0)
        {
            isPrime = false;
            break;
        }
    }
    if (isPrime)
    {
        Console.WriteLine(i);
    }
}
```
