# เลขคู่ เลขคี่ เลขเฉพาะ

```csharp

    {
        // แสดงเลขคี่
        System.Console.WriteLine("เลขคี่ (Odd numbers) จาก 1 ถึง 100:");
        for (int i = 1; i <= 100; i++)
        {
            if (i % 2 != 0)
            {
                System.Console.Write(i + " ");
            }
        }

        System.Console.WriteLine("\n");

        // แสดงเลขคู่
        System.Console.WriteLine("เลขคู่ (Even numbers) จาก 1 ถึง 100:");
        for (int i = 1; i <= 100; i++)
        {
            if (i % 2 == 0)
            {
                System.Console.Write(i + " ");
            }
        }

        System.Console.WriteLine("\n");

        // แสดงจำนวนเฉพาะ
        System.Console.WriteLine("จำนวนเฉพาะ (Prime numbers) จาก 1 ถึง 100:");
        for (int i = 2; i <= 100; i++)
        {
            if (IsPrime(i))
            {
                System.Console.Write(i + " ");
            }
        }

        System.Console.WriteLine(); // จบบรรทัด
    }

    static bool IsPrime(int number)
    {
        if (number < 2)
        {
            return false;
        }

        for (int i = 2; i * i <= number; i++)
        {
            if (number % i == 0)
            {
                return false;
            }
        }

        return true;
    }
```
