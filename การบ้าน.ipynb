```
// File: OddNumbers.cs
using System;

public class OddNumbers
{
    public static void Main(string[] args)
    {
        Console.WriteLine("--- เลขคี่ 1 - 100 ---");

        for (int i = 1; i <= 100; i++)
        {
            // ตรวจสอบว่าเป็นเลขคี่หรือไม่
            // ถ้าหารด้วย 2 แล้วเหลือเศษไม่เท่ากับ 0
            if (i % 2 != 0)
            {
                Console.Write(i + " ");
            }
        }
        Console.WriteLine(); // ขึ้นบรรทัดใหม่เมื่อจบ
    }
}
```

```
// File: EvenNumbers.cs
using System;

public class EvenNumbers
{
    public static void Main(string[] args)
    {
        Console.WriteLine("--- เลขคู่ 1 - 100 ---");

        for (int i = 1; i <= 100; i++)
        {
            // ตรวจสอบว่าเป็นเลขคู่หรือไม่
            // ถ้าหารด้วย 2 แล้วเหลือเศษเท่ากับ 0
            if (i % 2 == 0)
            {
                Console.Write(i + " ");
            }
        }
        Console.WriteLine(); // ขึ้นบรรทัดใหม่เมื่อจบ
    }
}
```

```
// File: PrimeNumbers.cs
using System;

public class PrimeNumbers
{
    public static void Main(string[] args)
    {
        Console.WriteLine("--- จำนวนเฉพาะ 1 - 100 ---");

        for (int i = 1; i <= 100; i++)
        {
            // เรียกใช้ฟังก์ชันเพื่อตรวจสอบว่าเป็นจำนวนเฉพาะหรือไม่
            if (IsPrime(i))
            {
                Console.Write(i + " ");
            }
        }
        Console.WriteLine(); // ขึ้นบรรทัดใหม่เมื่อจบ
    }

    /// <summary>
    /// ฟังก์ชันตรวจสอบจำนวนเฉพาะ
    /// </summary>
    /// <param name="number">ตัวเลขที่ต้องการตรวจสอบ</param>
    /// <returns>true ถ้าเป็นจำนวนเฉพาะ, false ถ้าไม่</returns>
    public static bool IsPrime(int number)
    {
        // 1 และตัวเลขที่น้อยกว่าไม่ถือเป็นจำนวนเฉพาะ
        if (number <= 1)
        {
            return false;
        }

        // วนลูปเช็คตัวหารตั้งแต่ 2 ถึงรากที่สองของตัวเลขนั้นๆ
        // การเช็คถึงแค่รากที่สอง (Math.Sqrt) จะช่วยให้โปรแกรมทำงานเร็วขึ้น
        for (int i = 2; i * i <= number; i++)
        {
            // ถ้าเจอตัวหารที่หารลงตัว แสดงว่าไม่ใช่จำนวนเฉพาะ
            if (number % i == 0)
            {
                return false;
            }
        }

        // ถ้าไม่เจอตัวหารใดๆ เลย แสดงว่าเป็นจำนวนเฉพาะ
        return true;
    }
}
```

```
using System;

public class AgeCategorizer
{
    public static void Main(string[] args)
    {
        Console.WriteLine("กรุณาป้อนอายุของคุณ:");
        string input = Console.ReadLine();
        int age;

        // ตรวจสอบว่าค่าที่ป้อนเข้ามาเป็นตัวเลขที่ถูกต้องหรือไม่
        if (int.TryParse(input, out age))
        {
            if (age >= 1 && age <= 12)
            {
                Console.WriteLine("คุณอยู่ในกลุ่ม: เด็ก");
            }
            else if (age >= 13 && age <= 19)
            {
                Console.WriteLine("คุณอยู่ในกลุ่ม: วัยรุ่น");
            }
            else if (age >= 20 && age <= 50)
            {
                Console.WriteLine("คุณอยู่ในกลุ่ม: วัยผู้ใหญ่");
            }
            else if (age >= 51)
            {
                Console.WriteLine("คุณอยู่ในกลุ่ม: วัยชรา");
            }
            else
            {
                Console.WriteLine("อายุที่คุณป้อนไม่ถูกต้อง กรุณาป้อนตัวเลขที่เป็นบวก");
            }
        }
        else
        {
            Console.WriteLine("ข้อมูลที่ป้อนไม่ถูกต้อง กรุณาป้อนเป็นตัวเลขเท่านั้น");
        }

        Console.WriteLine("\nกดปุ่มใดๆ เพื่อออก...");
        Console.ReadKey();
    }
}
```
