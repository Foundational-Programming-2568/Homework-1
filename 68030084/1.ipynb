```csharp
Console.WriteLine("เลขคี่ 1-100:");
for (int i = 1; i <= 100; i++)
{
    if (i % 2 != 0)
        Console.Write($"{i} ");
}
Console.WriteLine("\n");

// แสดงเลขคู่ระหว่าง 1-100
Console.WriteLine("เลขคู่ 1-100:");
for (int i = 1; i <= 100; i++)
{
    if (i % 2 == 0)
        Console.Write($"{i} ");
}
Console.WriteLine("\n");

// แสดงจำนวนเฉพาะระหว่าง 1-100
Console.WriteLine("จำนวนเฉพาะ 1-100:");
for (int i = 2; i <= 100; i++)
{
    bool isPrime = true;
    for (int j = 2; j <= Math.Sqrt(i); j++)
    {
        if (i % j == 0)
        {
            isPrime = false;
            break;
        }
    }
    if (isPrime)
        Console.Write($"{i} ");
}
Console.WriteLine();
```
